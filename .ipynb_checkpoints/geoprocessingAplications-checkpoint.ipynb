import shp files using geopandas

```python
SA1 = gpd.read_file('Geoprocessingfiles/Study_Area_1.shp')
SA2 = gpd.read_file('Geoprocessingfiles/Study_Area_2.shp')
river = gpd.read_file('Geoprocessingfiles/river.shp')
us_airports = pd.read_csv('Geoprocessingfiles/us_airports.csv')
us_states = gpd.read_file('Geoprocessingfiles/us_states.shp')
```

determining polygons intersection

```python
intersections = gpd.overlay(SA1, SA2, how='intersection')
intersections.plot(color='blue', figsize=(10, 10), alpha=0.3)
```

```python
intersections
```

determining polygons union

```python
union = gpd.overlay(SA1, SA2, how='union')
union.plot(color='blue', figsize=(10, 10), alpha=0.3, edgecolor='black')
```

```python
union
```

determining polygons symmetric difference

```python
sd = gpd.overlay(SA1, SA2, how='symmetric_difference')
sd.plot(color='blue', figsize=(10, 10), alpha=0.3, edgecolor='black')
```

```python
sd
```

determining polygons difference

```python
diff_sa1_sa2 = gpd.overlay(SA1, SA2, how='difference')
diff_sa1_sa2.plot(color='blue', figsize=(10, 10), alpha=0.3, edgecolor='black')
```

```python
diff_sa2_sa1 = gpd.overlay(SA2, SA1, how='difference')
diff_sa2_sa1.plot(color='blue', figsize=(10, 10), alpha=0.3, edgecolor='black')
```

determine polygons' dissolve

```python
union = gpd.overlay(SA1, SA2, how='union')
union['common_column'] = True
dissolved = union.dissolve(by='common_column')
dissolved.plot(color='blue', figsize=(10, 10), alpha=0.3, edgecolor='black')
```

```python
dissolved
```

determine polygons' centroids

```python
union = gpd.overlay(SA1, SA2, how='union')
union = union.to_crs(epsg = 24547)
centroid = union['geometry'].centroid
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)
union.plot(color='blue', edgecolor='black', ax=ax, alpha=0.3)
centroid.plot(ax=ax, color='black')
```

create a geometrical coordinate from an xy coordinate

```python
geometry = [Point(xy) for xy in zip(us_airports['LONGITUDE'], us_airports['LATITUDE'])]
geom_us_airports = gpd.GeoDataFrame(us_airports, geometry=geometry, crs=4326)
```

```python
geom_us_airports.plot(figsize=(10, 10))
```

perform geospatial joins

```python
geometry = [Point(xy) for xy in zip(us_airports['LONGITUDE'], us_airports['LATITUDE'])]
geom_us_airports = gpd.GeoDataFrame(us_airports, geometry=geometry, crs=4326)
geom_us_airports = gpd.sjoin(geom_us_airports, us_states, how='left', predicate='intersects')
geom_us_airports.drop(columns=['index_right'], inplace=True)
geom_us_airports.rename(columns={'NAME_1':'STATE NAME'}, inplace=True)
geom_us_airports.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 20)
geom_us_airports.plot(cmap='jet', column='STATE NAME', ax=ax, markersize=2)
us_states.plot(cmap='jet', column='NAME_1', ax=ax, alpha=0.3, legend=True)
```
